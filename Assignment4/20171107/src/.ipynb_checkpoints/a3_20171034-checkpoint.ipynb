```python
import scipy
import numpy as np
from scipy import ndimage
from scipy import misc
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
finger=cv.imread("fingerprint.jpg",0)

_,bin_finger=cv.threshold(finger,100,1,cv.THRESH_BINARY)
print(bin_finger)
plt.imshow(bin_finger,cmap='gray')
plt.show()
def hits(img,s):
    a,b=img.shape
    output=np.ones((a,b),dtype='uint8')
    #A=np.zeros(3,3)
    for i in range(1,a-1):
        for j in range(1,b-1):
            flag=1
            for p in range(0,3):
                for q in range(0,3):
                    if(s[p][q]!=img[i+p-1][j+q-1] and (s[p][q]==1 or s[p][q]==0)):
                        flag=0
            if(flag==1):
                output[i][j]=0
            else:
                output[i][j]=1
    return output

s1=np.asarray([[0,0,0],[2,1,2],[1,1,1]])
s2=np.asarray([[2,0,0],[1,1,0],[2,1,2]])
im1=hits(bin_finger,s1)
#plt.imshow(im1,cmap='gray')
#plt.show()

im2=hits(bin_finger,s2)
#plt.imshow(im2,cmap='gray')
#plt.show()

def logical_orr(img1,img2):
    a,b=img1.shape
    op=np.ones((a,b),dtype='uint8')
    for i in range(a):
        for j in range(b):
            if(img1[i][j]==1 and img2[i][j]==1):
                op[i][j]=1
            else:
                op[i][j]=0
    return op
#for i in range(im1.shape[0]):
#    for j in range(im1.shape[1]):
#        if(im2[i][j]==1):
#            print("yes")
def skele(img):
    
    s11= ndimage.rotate(s1, 90)
    s12= ndimage.rotate(s2, 90)
    s21= ndimage.rotate(s1, 180)
    s22= ndimage.rotate(s2, 180)
    s31= ndimage.rotate(s1, 270)
    s32= ndimage.rotate(s2, 270)
    im2=hits(img,s2)
    im3=hits(img,s11)
    im4=hits(img,s12)
    im5=hits(img,s21)
    im6=hits(img,s22)
    im7=hits(img,s31)
    im8=hits(img,s32)

    ans=logical_orr(im1,im2)
#plt.imshow(ans,cmap='gray')
#plt.show()

    ans=logical_orr(ans,im3)
    ans=logical_orr(ans,im4)
    ans=logical_orr(ans,im5)
    ans=logical_orr(ans,im6)
    ans=logical_orr(ans,im7)
    ans=logical_orr(ans,im8)
    plt.imshow(ans,cmap='gray')
    plt.show()
    return ans

ans=skele(bin_finger)
#print(s11)
```

```python
ans=skele(ans)
ans=skele(ans)
ans=skele(ans)
ans=skele(ans)
ans=skele(ans)
```
