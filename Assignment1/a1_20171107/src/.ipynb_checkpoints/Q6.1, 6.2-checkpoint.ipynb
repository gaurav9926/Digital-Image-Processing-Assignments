```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
def histEqualization(img):
    img = np.array(img)
#     img = np.array(img)
    img_flat = img.flatten()
    # plt.hist(img_flat, bins=50)
    # print()
    # print(img_flat)

    histogram = np.zeros(256)
    for pixel in img_flat:
        histogram[pixel] += 1
    for i in range(1,256):
    #     print(i)
        histogram[i] += histogram[i-1]
    histogram = (((histogram-np.min(histogram))/(np.max(histogram)-np.min(histogram))) *255).astype('uint8') 
    r, c = img.shape
    # prin
    # histogram = np.floor((histogram/r*c)*255)
    img_new_flat = histogram[img_flat]
    # plt.hist(img_new, bins=50)
    img_new = np.reshape(img_new_flat, img.shape)
    fig = plt.figure(figsize=[10,5]);
    sub1 = fig.add_subplot(1,2,1);
    # sub1.imshow(img, cmap='gray')
    sub1.hist(img_flat, bins=256)
    sub2 = fig.add_subplot(1,2,2)
    # sub2.imshow(img_new, cmap='gray');plt.axis('off')
    sub2.hist(img_new_flat, bins=256)
    plt.show()
    return img_new
img = Image.open('../input/lena.jpg').convert('L')
img_new = histEqualization(img)
# plt.imshow(img_new, cmap='gray')
```

```python
fig = plt.figure(figsize=[10,5]);
sub1 = fig.add_subplot(1,2,1);
sub1.imshow(img, cmap='gray');sub1.axis('off')
# sub1.hist(img_flat, bins=256)
sub2 = fig.add_subplot(1,2,2)
sub2.imshow(img_new, cmap='gray');sub2.axis('off')
# sub2.hist(img_new_flat, bins=256)
plt.show()
```
