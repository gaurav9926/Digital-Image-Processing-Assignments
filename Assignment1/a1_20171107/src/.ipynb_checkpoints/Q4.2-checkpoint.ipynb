```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2, math, string
from fractions import Fraction
def gammaFunct(img, gamma):
    img = np.array(img)
    img = (img/255)**gamma
#     type(img)
#     gamma = Fraction(gamma)
#     img1 = (img/255)**gamma
#     print(img1)
#     return Image.fromarray(img)
    return img
img = Image.open('../input/gamma-corr.png')
img = np.array(img)
k = 0.3
# k = 1
image = gammaFunct(img, k)
fig = plt.figure(figsize=[10,10]);
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.imshow(img, cmap='gray');ax1.axis('off')
ax2.imshow(image,cmap='gray');ax2.axis('off')

image1 = gammaFunct(img, 1)
fig = plt.figure(figsize=[10,10]);
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.imshow(img, cmap='gray');ax1.axis('off')
ax2.imshow(image1,cmap='gray');ax2.axis('off')

image2 = gammaFunct(img, 2)
fig = plt.figure(figsize=[10,10]);
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.imshow(img, cmap='gray');ax1.axis('off')
ax2.imshow(image2,cmap='gray');ax2.axis('off')

plt.show()
# for gamma in range(0,3):
#     k = gamma+0.5
#     image = gammaFunct(img, gamma)
#     fig = plt.figure(figsize=[10,10]);
#     ax1 = fig.add_subplot(1,2,1)
#     ax2 = fig.add_subplot(1,2,2)
#     ax1.imshow(img, cmap='gray');ax1.axis('off')
#     ax2.imshow(image,cmap='gray');ax2.axis('off')
#     plt.show()
```
