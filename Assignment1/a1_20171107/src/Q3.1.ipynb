```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
def BitQuantizeImage(img, k):
    img = np.array(img)
    img = img/256
    img = np.floor(img*(2**k)).astype('uint8')
#     print(img)
    return (img/2**k)
img = Image.open('../input/quantize.jpg').convert('RGB')
# to display Images
for i in range(1,9):
    new_im = BitQuantizeImage(img,i)
    plt.imshow(new_im)
    plt.axis('off')
    plt.show()
# image = BitQuantizeImage(img, k)
# plt.imshow(image/(2**i)),plt.axis('off')
# plt.show()
```
