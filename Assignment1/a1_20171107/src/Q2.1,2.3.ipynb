```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
def linContrastStretching(img, a, b):
    img = np.array(img)
    img1= img.copy()
    minval = np.min(img1)
    maxval = np.max(img1)
    mi = np.min(img1[:][:])
    ma = np.max(img1[:][:])
    x, y = img1.shape
    for i in range(0,x):
        for j in range(0,y):
            if img1[i][j] < mi:
                img1[i][j] = mi
            elif img1[i][j] > ma:
                img1[i][j] = ma
            else:
                img1[i][j] = ((img1[i][j]-mi)/(ma-mi))*(b-a)
    img1 = img1.astype('uint8')
    fig = plt.figure(figsize=[10, 10])
    ax1 = fig.add_subplot(2,2,1)
    ax2 = fig.add_subplot(2,2,2)
    ax3 = fig.add_subplot(2,2,3)
    ax4 = fig.add_subplot(2,2,4)
    ax1.imshow(img, cmap='gray');ax1.axis('off')
    ax2.imshow(img1, cmap='gray');ax2.axis('off')
    ax3.hist(img.flatten(), bins=256)
    ax4.hist(img1.flatten(), bins=(b-a))
    
    return img1
img = Image.open('../input/cameraman.png').convert('L')
# plt.imshow(img, cmap='gray')
image = linContrastStretching(img,0,255)
# plt.imshow(image,cmap='gray')

```
