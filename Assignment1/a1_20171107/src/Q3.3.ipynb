```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
def BitQuantizeImage(img, k):
    img = np.array(img)
    img = img/256
    img = np.floor(img*(2**(k))).astype('uint8')
    return img

def bitPlane(img, k):
    img = np.array(img)
    img = np.mod(np.floor(img/2**k), 2).astype('uint8')
    return img 


img = Image.open('../input/lena.jpg').convert('L')
lena1img = Image.open('../input/lena1.jpg').convert('L')
lena2img = Image.open('../input/lena2.jpg').convert('L')
lena3img = Image.open('../input/lena3.jpg').convert('L')
# new_img1 = BitQuantizeImage(img, 5)  
new_img1 = bitPlane(img, 4)
new_img2 = BitQuantizeImage(img, 2)
new_img3 = BitQuantizeImage(img, 1)

fig = plt.figure(figsize=[30,30])

ax1 = fig.add_subplot(3,2,1)
ax1.imshow(lena1img, cmap='gray');ax1.axis('off')
ax2 = fig.add_subplot(3,2,2)
ax2.imshow(new_img1, cmap='gray');ax2.axis('off')

ax3 = fig.add_subplot(3,2,3)
ax3.imshow(lena2img, cmap='gray');ax3.axis('off')
ax4 = fig.add_subplot(3,2,4)
ax4.imshow(new_img2, cmap='gray');ax4.axis('off')

ax5 = fig.add_subplot(3,2,5)
ax5.imshow(lena3img, cmap='gray');ax5.axis('off')
ax6 = fig.add_subplot(3,2,6)
ax6.imshow(new_img3, cmap='gray');ax6.axis('off')

plt.show()
```

Lena2 Image is 2 Bit Quantisation of the image 'lena.jpg' <br>
Lena3 Image is 1 Bit Quantisation of the image 'lena.jpg' <br>
