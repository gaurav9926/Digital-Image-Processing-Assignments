```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
def bitPlane(img, k):
    img = np.array(img)
    img = np.mod(np.floor(img/2**(k-1)), 2).astype('uint8')
    return img 
img = Image.open('../input/lena.jpg').convert('L')
for i in range(1,9):
    new_im = bitPlane(img,i)
    plt.imshow(new_im, cmap='gray')
    plt.axis('off')
    plt.show()

# img = np.array(img)
# img1 = np.mod(np.floor(img/2), 2).astype('uint8')
# img1 = bitPlane(img, 4)
# plt.imshow(img1, cmap='gray')
```
