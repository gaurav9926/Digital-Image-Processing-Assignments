```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from math import sqrt
```

```python
def make_filt(img):
    filt = np.zeros(img.shape)
    filt = filt.astype('float')
    for u in range(img.shape[0]):
        for v in range(img.shape[1]):
            filt[u][v] = sqrt((u-((img.shape[0])/2))**2 + (v-((img.shape[1])/2))**2)
#     print(filt)
    return filt

def take_mult(img, filt):
    img = np.array(img)
    img = img.astype('float')
    fft_img = np.fft.fft2(img)
    fft_img_shift = np.fft.fftshift(fft_img)
    
    magnitude_spectrum = 20*np.log(1+np.abs(fft_img_shift))
#     fft_img_shift[h_rows -10:h_rows +10, h_cols -10:h_cols +10] = 0

    new_img = np.multiply(fft_img_shift, filt)
    magnitude_spectrum2 = 20*np.log(1+np.abs(new_img))

#   Making Image Back to Spatial Domain
    new_img_ifft = np.fft.ifftshift(new_img)
    img_back = np.fft.ifft2(new_img_ifft)
    img_back = np.abs(img_back)
    img_back = img_back.astype('uint8')
    
    figure = plt.figure(figsize=[15,15])
    plt.subplot(121),plt.imshow(magnitude_spectrum, cmap = 'gray')
    plt.title('Input Image'),plt.axis('off')
    plt.subplot(122),plt.imshow(magnitude_spectrum2, cmap = 'gray')
    plt.title('After Transform'),plt.axis('off')
    plt.show()
    
    return img_back
```

```python
def linearLOWPASS(img, d0):
    img = np.array(img)
    img = img.astype('float')
    rows, cols = img.shape
    
#     filt = np.ones((k,k)).astype('float')
#     filt = filt/(k*k)
    filt = np.zeros(img.shape)
    filt = filt.astype('float')
    temp = make_filt(img)
    for i in range(rows):
        for j in range(cols):
            if temp[i][j] <= d0:
                filt[i][j] = 1
            else:
                filt[i][j] = 0
    
#     fft_filt = np.fft.fft2(filt,img.shape)
#     fft_filt_shift = np.fft.fftshift(fft_filt)
#     take_mult(img, filt)
    return take_mult(img, filt)
```

```python
img = Image.open('../input_data/ImageProcessingDemo.png').convert('L')
img = np.array(img)
ans = linearLOWPASS(img, 100)

figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'),plt.axis('off')
plt.subplot(122),plt.imshow(ans, cmap = 'gray')
plt.title('After Transform'),plt.axis('off')
plt.show()
```

```python
def butterworthLOWPASS(img, d0, n):
    img = np.array(img)
    img = img.astype('float')
    
    temp = make_filt(img)
    temp = temp/d0
    temp = temp**(2*n)
    temp = 1 + temp
    filt = np.zeros(img.shape)
    filt = 1/temp
    
    return take_mult(img, filt)
    
```

```python
img = Image.open('../input_data/ImageProcessingDemo.png').convert('L')
img = np.array(img)
ans = butterworthLOWPASS(img, 50, 2)

figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'),plt.axis('off')
plt.subplot(122),plt.imshow(ans, cmap = 'gray')
plt.title('After Transform'),plt.axis('off')
plt.show()
```

```python
def gaussianLOWPASS(img, d0):
    temp = make_filt(img)
    filt = np.zeros(img.shape)
    filt = np.exp(-(np.square(temp))/(2*d0*d0))
    
    return take_mult(img, filt)
```

```python
img = Image.open('../input_data/ImageProcessingDemo.png').convert('L')
img = np.array(img)
ans = gaussianLOWPASS(img, 30)

figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'),plt.axis('off')
plt.subplot(122),plt.imshow(ans, cmap = 'gray')
plt.title('After Transform'),plt.axis('off')
plt.show()
```

```python
img = Image.open('../input_data/lena.jpg').convert('L')
img = np.array(img)
ans = gaussianLOWPASS(img, 30)
ans1 = gaussianLOWPASS(img, 60)
new_ans = np.zeros(img.shape)
new_ans = np.abs(ans-ans1)


figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(new_ans, cmap = 'gray')
plt.title('Difference'),plt.axis('off')
# plt.subplot(122),plt.imshow(ans1, cmap = 'gray')
# plt.title('After Transform'),plt.axis('off')
plt.show()
```

<b>As we increase the value of sigma the blur of the image increases. And when we take difference of two sigma value images, that range of frequencies become dominant.</b>
