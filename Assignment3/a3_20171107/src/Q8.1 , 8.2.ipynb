```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
def compute_spectrum(img):
    img = np.array(img)
    img = img.astype('float')
    fft_img = np.fft.fft2(img)
    fft_img_shift = np.fft.fftshift(fft_img)
    
    magnitude_spectrum = 20*np.log(1+np.abs(fft_img_shift))
    return magnitude_spectrum
def compute_phase_spectrum(img):
    img = np.array(img)
    img = img.astype('float')
    fft_img = np.fft.fft2(img)
    fft_img_shift = np.fft.fftshift(fft_img)

    phase_spectrum = np.angle(fft_img_shift)
    return phase_spectrum
```

```python
def translate_image(img, k):
    img = np.array(img)
    ans = np.zeros(img.shape)
    ans[:,0:k] = img[:,img.shape[1]-k:img.shape[1]]
    ans[:,k:img.shape[1]] = img[:,0:img.shape[1]-k]
    return ans
```

```python
X = [[1,2,3],[4,5,6],[7,8,9]]
translate_image(X, 2)
```

```python
img = Image.open('../input_data/rectangle.jpg')
ans = translate_image(img, 70)

magnitude_spectrum = compute_spectrum(img)
magnitude_spectrum1 = compute_spectrum(ans)

phase_spectrum = compute_phase_spectrum(img)
phase_spectrum1 = compute_phase_spectrum(ans)

figure = plt.figure(figsize=[20,20])
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'),plt.axis('off')
plt.subplot(122),plt.imshow(ans, cmap = 'gray')
plt.title('After Transform'),plt.axis('off')
plt.show()

figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Input Spectrum'),plt.axis('off')
plt.subplot(122),plt.imshow(magnitude_spectrum1, cmap = 'gray')
plt.title('Translated Spectrum'),plt.axis('off')

figure = plt.figure(figsize=[15,15])
plt.subplot(121),plt.imshow(phase_spectrum, cmap = 'gray')
plt.title('Input Phase Spectrum'),plt.axis('off')
plt.subplot(122),plt.imshow(phase_spectrum1, cmap = 'gray')
plt.title('Translated Phase Spectrum'),plt.axis('off')
plt.show()
```
