```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageOps
import numpy as np
import time
from math import exp,pi



def kernel(img,i,j,sigmad,sigmar,k):
    d = np.zeros((k,k))
    r = np.zeros((k,k))
    d = d.astype('float')
    r = r.astype('float')
    k1 = k//2
    for i in range(k):
        for j in range(k):
            d[i,j] = (i-k1)**2 + (j-k1)**2
    d = (1/(2*pi*(sigmad**2)))*(np.exp(-(d/(2*(sigmad**2)))))
    temp = img[i-k1:i+k1+1,j-k1:j+k1+1]
    r = temp - img[i,j]
    r = (1/(2*pi*(sigmar**2)))*(np.exp(-(np.square(r)/(2*(sigmar**2)))))
    return d*r

def CrossbilateralFilter(img1 ,img2, sigmad, sigmar):
    k = 11
    k1 = k//2
    ans = np.array(img1)
    ker = np.zeros((k,k))
    ker = ker.astype('float')
    img1 = img1.astype('float')
    img2 = img2.astype('float')
    ans = ans.astype('float')
    x , y = img1.shape
    sumi = float(0)
    temp1 = float(0)
   
    for i in range(k1,x-k1):
            for j in range(k1,y-k1):
                ker = kernel(img1,i,j,sigmad,sigmar,k)
                temp1 = img2[i-k1:i+k1+1,j-k1:j+k1+1]*ker
                sumi = np.sum(ker)
                ans[i][j] = np.sum(temp1/sumi)
    ans[ans>255] = 255
    ans[ans<0] = 0
    return ans.astype('uint8')
def CrossbilateralFilter3D(img1 ,img2 ,sigmad ,sigmar):
    ans = np.zeros(img1.shape)
#     print(img1[:,:,0].shape)
    ans[:,:,0] = CrossbilateralFilter(img1[:,:,0],img2[:,:,0],sigmad,sigmar)
    ans[:,:,1] = CrossbilateralFilter(img1[:,:,1],img2[:,:,1],sigmad,sigmar)    
    ans[:,:,2] = CrossbilateralFilter(img1[:,:,2],img2[:,:,2],sigmad,sigmar)    
    return ans


```

```python
# img = Image.open('sky.png').convert('L')
img1 = cv2.imread('../input/pots_flash.jpg',-1)
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
# print(img1.shape)
img2 = cv2.imread('../input/pots_no_flash.jpg',-1)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
# print(img2.shape)
ans = CrossbilateralFilter3D(img1,img2, 40, 0.05)

fig = plt.figure(figsize=[15,15])
plt.subplot(1,3,1);plt.imshow(img1);plt.axis('off')
plt.subplot(1,3,2);plt.imshow(img2);plt.axis('off')
plt.subplot(1,3,3);plt.imshow(ans.astype('uint8'))
plt.show()
```

```python
def InvbilateralFilter(img, sigmad, sigmar):
    k = 5
    k1 = k//2
    ans = np.array(img)
    ker = np.zeros((k,k))
    ker = ker.astype('float')
    img = img.astype('float')
    img = (img-np.min(img))/(np.max(img)-np.min(img))
    ans = ans.astype('float')
    x , y = img.shape
    sumi = float(0)
    temp1 = float(0)
   
    for i in range(k1,x-k1):
            for j in range(k1,y-k1):
                ker = kernel(img,i,j,sigmad,sigmar,k)
                temp1 = img[i-k1:i+k1+1,j-k1:j+k1+1]*ker
                sumi = np.sum(ker)
                ans[i][j] = np.sum(temp1/sumi)
    ans = ans*255
    ans[ans>255] = 255
    ans[ans<0] = 0
    return ans.astype('uint8')
def InvbilateralFilter3D(img, sigmad, sigmar):
    ans = np.zeros(img.shape)
    ans[:,:,0] = InvbilateralFilter(img[:,:,0],sigmad,sigmar)
    ans[:,:,1] = InvbilateralFilter(img[:,:,1],sigmad,sigmar)    
    ans[:,:,2] = InvbilateralFilter(img[:,:,2],sigmad,sigmar)    
    return ans
```

```python
img = cv2.imread('../input/bell.jpg',-1)
# img1 = cv2.imread('sky.png',0)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
ans = InvbilateralFilter3D(img, 20,20)
# ans1 = InvbilateralFilter(img1,20,20)
# plt.imshow(ans)
# print(ans)
fig = plt.figure(figsize=[10,10])
plt.subplot(1,2,1);plt.imshow(img);plt.axis('off')
plt.subplot(1,2,2);plt.imshow(ans.astype('uint8'));plt.axis('off')
plt.show()
```
