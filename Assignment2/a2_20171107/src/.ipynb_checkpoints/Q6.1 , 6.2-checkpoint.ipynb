```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageOps
import numpy as np
import time
from math import pi

def kernel(img,i,j,sigmad,sigmar,k):
    d = np.zeros((k,k))
    r = np.zeros((k,k))
    d = d.astype('float')
    r = r.astype('float')
    k1 = k//2
    for i in range(k):
        for j in range(k):
            d[i,j] = (i-k1)**2 + (j-k1)**2
    d = (1/(2*pi*(sigmad**2)))*np.exp(-(d/(2*(sigmad**2))))
    temp = img[i-k1:i+k1+1,j-k1:j+k1+1]
    r = temp - img[i,j]
    r = (1/(2*pi*(sigmar**2)))*np.exp(-(np.square(r))/(2*(sigmar**2)))
    return d*r

def bilateralFilter(img, sigmad, sigmar):
    k = 5
    k1 = k//2
    ans = np.array(img)
    ker = np.zeros((k,k))
    ker = ker.astype('float')
    img = img.astype('float')
    ans = ans.astype('float')
    x , y = img.shape
    sumi = float(0)
    temp1 = float(0)
   
    for i in range(k1,x-k1):
            for j in range(k1,y-k1):
                ker = kernel(img,i,j,sigmad,sigmar,k)
                temp1 = img[i-k1:i+k1+1,j-k1:j+k1+1]*ker
                sumi = np.sum(ker)
                ans[i][j] = np.sum(temp1/sumi)
    ans[ans>255] = 255
    ans[ans<0] = 0
    return ans.astype('uint8')
def bilateralFilter3D(img, sigmad, sigmar):
    ans = np.array(img)
#     ans = np.array(img)
    ans[:,:,0] = bilateralFilter(img[:,:,0],sigmad,sigmar)
    ans[:,:,1] = bilateralFilter(img[:,:,1],sigmad,sigmar)    
    ans[:,:,2] = bilateralFilter(img[:,:,2],sigmad,sigmar)    
    return ans
```

```python
img = cv2.imread('../input/sky.png',-1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
ans = bilateralFilter3D(img,50,10)
fig = plt.figure(figsize=[20,20])
plt.subplot(1,2,1);plt.imshow(img)
plt.subplot(1,2,2);plt.imshow(ans)
plt.show()
```

```python
def error(img,img1,sigmad,sigmar):
    ans = bilateralFilter3D(img,sigmad,sigmar)
    error = np.sqrt(np.sum(((ans.astype('float')-img1.astype('float'))**2)))
    error = round(error,3)
#     error = error.astype('uint8')
#     print(error)
    fig = plt.figure(figsize=[5,5])
#     pl.subplot(121)
    plt.imshow(ans);plt.axis('off');plt.title('Sigmad={}, Sigmar={}, L2error={}'.format(sigmad,sigmar,error))
    plt.show()

    
img = cv2.imread('../input/noir.png',-1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img1 = cv2.imread('../input/gt_noir.png',-1)
img = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

error(img,img1,0.5,0.5)
error(img,img1,3.5,0.2)
error(img,img1,3,3)
error(img,img1,10,10)
error(img,img1,15,15)
error(img,img1,30,30)
```

```python
img = cv2.imread('../input/sky.png',-1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img1 = cv2.imread('../input/gt_sky.png',-1)
img = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

error(img,img1,0.5,0.5)
error(img,img1,3.5,0.2)
error(img,img1,3,3)
error(img,img1,10,10)
error(img,img1,15,15)
error(img,img1,30,30)
```
